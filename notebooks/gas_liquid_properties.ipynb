# Material properties of water, DCM, and some gases

## Dependencies

This notebook has no dependencies to other notebooks or to the data set. The literature sources for the material properties used below can be found in the appendix of my thesis.

## Visualizations

```python
import helper_module as hm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

%matplotlib inline
rc('text', usetex=True)

marker_size = 12
line_width = 2
```

```python
# references for the numerical values can be found in the appenix (material properties of selected liquids and gases)
# of my PhD thesis

# material properties of water
eta_water_temp = np.asarray([1.0020, 0.8802, 0.6522, 0.2821, 0.2313]) * 1.0E-3
rho_water_temp = np.asarray([998.21, 997.04, 992.21, 958.35, 943.11])
nu_water_temp = np.asarray([1.0038, 0.8828, 0.6753, 0.2944, 0.2453]) * 1.0E-6
sigma_water_temp = np.asarray([72.750, 71.990, 69.600, 58.910, 54.960]) * 1.0E-3
temp_water = np.asarray([20.0, 25.0, 40.0, 100.0, 120.0])

# material properties of DCM
eta_dcm_temp = np.asarray([0.4350, 0.4130, 0.3970]) * 1.0E-3
rho_dcm_temp = np.asarray([1323.1, 1314.2, 1305.0])
nu_dcm_temp = np.asarray([0.3287, 0.3143, 0.3042]) * 1.0E-6
sigma_dcm_temp = np.asarray([27.840, 27.198, 26.555]) * 1.0E-3
temp_dcm = np.asarray([20.0, 25.0, 40.0])

# material properties of ethylene gas in water
eta_ethylene_temp_1atm = np.asarray([1.015, 1.032, 1.082, 1.277, 1.339]) * 1.0E-5
rho_ethylene_temp_1atm = np.asarray([1.157, 1.138, 1.083, 0.906, 0.860])
nu_ethylene_temp_1atm = np.asarray([0.877, 0.907, 0.999, 1.409, 1.557]) * 1.0E-5
H_ethylene_temp_1atm = np.asarray([6.132, 6.838, 9.271, 24.08, 30.84])
D_ethylene_temp_1atm = np.asarray([1.398, 1.590, 2.283, 7.256, 9.863]) * 1.0E-5
eta_ethylene_temp_4atm = np.asarray([1.022, 1.039, 1.089, 1.286, 1.346]) * 1.0E-5
rho_ethylene_temp_4atm = np.asarray([4.887, 4.773, 4.475, 3.702, 3.531])
nu_ethylene_temp_4atm = np.asarray([0.240, 0.218, 0.243, 0.347, 0.381]) * 1.0E-5

# material properties of several gases
# order:
# argon, carbon-dioxide, hydrogen, nitrogen, nitric oxide, oxygen
labels = [r"$A$", r"$CO_2$", r"$H_2$", r"$N_2$", r"$NO$", r"$O_2$"]
eta_gases = np.asarray([2.23, 1.47, 0.88, 1.76, 1.47, 2.04]) * 1.0E-5
rho_gases = np.asarray([1.661, 1.842, 0.083, 1.165, 1.249, 1.331])
nu_gases = np.asarray([1.343, 0.798, 10.60, 1.511, 1.177, 1.533]) * 1.0E-5
H_gases = np.asarray([29.31, 1.243, 52.60, 64.11, 21.59, 34.19])
```

```python
# plot properties of water
fig, ax = plt.subplots(figsize=(hm.figure_width, 6))

ax.plot(temp_water, eta_water_temp/eta_water_temp[0], marker='*', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\eta/\eta_{20}$")
ax.plot(temp_water, rho_water_temp/rho_water_temp[0], marker='<', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\rho/\rho_{20}$")
ax.plot(temp_water, nu_water_temp/nu_water_temp[0], marker='o', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\nu/\nu_{20}$")
ax.plot(temp_water, sigma_water_temp/sigma_water_temp[0], marker='x', ms=marker_size,
        c='k', ls='--', lw=line_width, label=r"$\sigma/\sigma_{20}$")

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=4,
           ncol=4, mode="expand", borderaxespad=0., fontsize=hm.fontsize_legend)
ax.set_xlabel(r"$\vartheta$ in $^\circ C$", fontsize=hm.fontsize_label)
ax.set_xticks([20, 25, 40, 60, 80, 100, 120])
ax.tick_params(labelsize=hm.fontsize_tick)
ax.grid(linestyle='--')

plt.savefig("../output/water_properties_vs_temperature.pdf", bbox_inches='tight')
plt.savefig("../output/water_properties_vs_temperature.png", bbox_inches='tight')
```

```python
# plot properties of DCM
fig, ax = plt.subplots(figsize=(hm.figure_width, 6))

ax.plot(temp_dcm, eta_dcm_temp/eta_dcm_temp[0], marker='*', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\eta/\eta_{20}$")
ax.plot(temp_dcm, rho_dcm_temp/rho_dcm_temp[0], marker='<', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\rho/\rho_{20}$")
ax.plot(temp_dcm, nu_dcm_temp/nu_dcm_temp[0], marker='o', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\nu/\nu_{20}$")
ax.plot(temp_dcm, sigma_dcm_temp/sigma_dcm_temp[0], marker='x', ms=marker_size,
        c='k', ls='--', lw=line_width, label=r"$\sigma/\sigma_{20}$")

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=4,
           ncol=4, mode="expand", borderaxespad=0., fontsize=hm.fontsize_legend)
ax.set_xlabel(r"$\vartheta$ in $^\circ C$", fontsize=hm.fontsize_label)
ax.set_xticks([20, 25, 40])
ax.tick_params(labelsize=hm.fontsize_tick)
ax.grid(linestyle='--')
plt.savefig("../output/dcm_properties_vs_temperature.pdf", bbox_inches='tight')
plt.savefig("../output/dcm_properties_vs_temperature.png", bbox_inches='tight')
```

```python
# plot properties of ethylene gas
fig, ax = plt.subplots(figsize=(hm.figure_width, 6))

ax.plot(temp_water, eta_ethylene_temp_1atm/eta_ethylene_temp_1atm[0], marker='*', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\eta/\eta_{20}$")
ax.plot(temp_water, rho_ethylene_temp_1atm/rho_ethylene_temp_1atm[0], marker='<', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\rho/\rho_{20}$")
ax.plot(temp_water, nu_ethylene_temp_1atm/nu_ethylene_temp_1atm[0], marker='o', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\nu/\nu_{20}$")
ax.plot(temp_water, H_ethylene_temp_1atm/H_ethylene_temp_1atm[0], marker='x', ms=marker_size,
        c='k', ls='--', lw=line_width, label=r"$H/H_{20}$")
ax.plot(temp_water, D_ethylene_temp_1atm/D_ethylene_temp_1atm[0], marker='^', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$D/D_{20}$")

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=4,
           ncol=5, mode="expand", borderaxespad=0., fontsize=hm.fontsize_legend)
ax.set_xlabel(r"$\vartheta$ in $^\circ C$", fontsize=hm.fontsize_label)
ax.set_xticks([20, 25, 40, 60, 80, 100, 120])
ax.tick_params(labelsize=hm.fontsize_tick)
ax.grid(linestyle='--')
plt.savefig("../output/water_ethylene_vs_temperature.pdf", bbox_inches='tight')
plt.savefig("../output/water_ethylene_vs_temperature.png", bbox_inches='tight')
```

```python
# plot properties of ethylene at 4 atm
fig, ax = plt.subplots(figsize=(hm.figure_width, 6))

ax.plot(temp_water, eta_ethylene_temp_1atm/eta_ethylene_temp_1atm[0], marker='*', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\eta/\eta_{20}$")
ax.plot(temp_water, rho_ethylene_temp_1atm/rho_ethylene_temp_1atm[0], marker='<', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\rho/\rho_{20}$")
ax.plot(temp_water, nu_ethylene_temp_1atm/nu_ethylene_temp_1atm[0], marker='o', ms=marker_size, markerfacecolor='none',
        c='k', ls='--', lw=line_width, label=r"$\nu/\nu_{20}$")

ax.plot(temp_water, eta_ethylene_temp_4atm/eta_ethylene_temp_1atm[0], marker='*', ms=marker_size, markerfacecolor='none',
        c='k', ls=':', lw=line_width, label=r"$\eta/\eta_{20}$ at $4\times10^5Pa$")
ax.plot(temp_water, rho_ethylene_temp_4atm/rho_ethylene_temp_1atm[0], marker='<', ms=marker_size, markerfacecolor='none',
        c='k', ls=':', lw=line_width, label=r"$\rho/\rho_{20}$ at $4\times10^5Pa$")
ax.plot(temp_water, nu_ethylene_temp_4atm/nu_ethylene_temp_1atm[0], marker='o', ms=marker_size, markerfacecolor='none',
        c='k', ls=':', lw=line_width, label=r"$\nu/\nu_{20}$ at $4\times10^5Pa$")

handles, labels = ax.get_legend_handles_labels()
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
plt.legend(handles, labels, bbox_to_anchor=(0., 1.02, 1., .102), loc=4,
           ncol=3, mode="expand", borderaxespad=0., fontsize=hm.fontsize_legend)
ax.set_xlabel(r"$\vartheta$ in $^\circ C$", fontsize=hm.fontsize_label)
ax.set_xticks([20, 25, 40, 60, 80, 100, 120])
ax.tick_params(labelsize=hm.fontsize_tick)
ax.grid(linestyle='--')
plt.savefig("../output/water_ethylene_vs_temperature_4atm.pdf", bbox_inches='tight')
plt.savefig("../output/water_ethylene_vs_temperature_4atm.png", bbox_inches='tight')
```
